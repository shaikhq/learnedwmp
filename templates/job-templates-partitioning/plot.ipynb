```python
import matplotlib.pyplot as plt
import pandas as pd

# MAPE scores collected across different seeds
data = {
    'seed': [123, 234, 456, 567, 678],
    'DBMS': [169.09, 57.12, 608.05, 677.36, 833.07],
    'LearnedWMP': [52.53, 46.95, 28.80, 31.29, 40.733]
}

df = pd.DataFrame(data)

# Create x positions for plotting
x = list(range(len(df)))
x_labels = [f"Seed {s}" for s in df['seed']]

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, df['DBMS'], marker='o', label='DBMS', linewidth=2)
plt.plot(x, df['LearnedWMP'], marker='o', label='LearnedWMP', linewidth=2)

# Annotate MAPE values above each point
for i in x:
    plt.text(x[i], df['DBMS'][i] + 20, f"{df['DBMS'][i]:.1f}%", ha='center', fontsize=8, color='blue')
    plt.text(x[i], df['LearnedWMP'][i] + 20, f"{df['LearnedWMP'][i]:.1f}%", ha='center', fontsize=8, color='green')

# Customize x-axis to show seed values
plt.xticks(ticks=x, labels=x_labels)

# Add labels and title
plt.title("MAPE Scores from 5 Random Train/Test Splits (No Template Overlap)")
plt.xlabel("Random Seed")
plt.ylabel("MAPE (%)")

# Remove all borders
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# No grid
# plt.grid(False)

plt.legend()
plt.tight_layout()
plt.savefig("mape_seed_comparison.png", dpi=300)  # Save as PNG with high resolution
plt.show()
```
