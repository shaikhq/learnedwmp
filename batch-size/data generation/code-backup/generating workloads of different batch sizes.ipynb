```python
def create_workload(batch_size, data, k):
    # keeping only the columns that I need
    df_data = data[['db2','actual','cluster']]

    # 1-hot encode the cluster column
    df_data = pd.get_dummies(df_data, columns=['cluster'])
    
    # create a dataframe of workload batches
    df_batches = pd.DataFrame(columns=df_data.columns)

    # debug - check the output from np.linspace
    print(data.shape[0] - 1)
    print(data.shape[0])
    
    # the following line generate a list of integers between the range of 0 and data.shape[0]-1. It generates data.shape[0] number of integers. 
    # If the dataframe has 1000 records, the value of data.shape[0] is 1000. Now we'll be generating 1000 numbers in the range (0, 999).
    # data.shape[0] - 1 = 999. The 3rd parameter tells the function to generate 1000 numbers. 
    # the final parameter is the datatype of the output numbers
    indices = np.linspace(0, data.shape[0]-1, data.shape[0], dtype=int)

    # shuffle these numbers
    np.random.shuffle(indices)
    
    num_batches = int(np.floor(df_data.shape[0] / batch_size))
        
    for ibat in range(num_batches):
        start = (ibat * batch_size)
        end = (ibat * batch_size + batch_size) - 1
        print("start: ", start, "end:", end)
        
        ibat_Y = df_data.loc[indices[start:end], :]

        #df_batches = pd.concat([df_batches, ibat_Y.sum()], ignore_index=True)
        
        df_batches = df_batches.append(ibat_Y.sum(), ignore_index=True)
        
    return df_batches
```
